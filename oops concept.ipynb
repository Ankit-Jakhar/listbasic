```python
#creating a calculator
class calculator:
     def __init__(self , num1 , num2):
         self.num1 =num1
         self.num2 =num2
     def add(self):
         return self.num1+self.num2
     def subtraction(self):
         return self.num1-self.num2
     def multiply(self):
         return self.num1*self.num2
     def divide(self):
          if self.num2!=0:
              return self.num1//self.num2
          else:
              return "division by zero is not allowed"
#test the calculator
calc  =calculator(10,5)
print(calc.add())
print(calc.subtraction())
print(calc.multiply())
print(calc.divide())
```
