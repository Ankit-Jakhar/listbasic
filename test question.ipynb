```python

##Q.11 What is view vs copy ? explain it with example ? 
#view ---->#view method mean modification in original array
import numpy as np
a =np.array([1,2,3,4,5,5,6,75,8]) 
b =a[2:5]
b[:]=0
b
a
```

```python
#copy --->#copy method will not update youe original array
a =np.array([1,2,3,4,5,6,6,76,8]) 
b =a[1:5].copy()
b[:]=0
b
a
```
